```python
import pandas as pd
```

```python
df=pd.read_csv("1-200/1-200（0303）.csv",encoding="gbk")
df[df.PIC_ID==1]
# labels={0:'walk',1:'T',2:'L',3:'car',4:'bike',5:'cover',6:'column'}
mydic={"行进盲道":0,"T型盲道":1,"L型盲道":2,"机动车":3,"非机动车":4,"窨井盖":5,"柱":6}
```

```python
import re

for pic_id in range(1,201):
    pic=df[df.PIC_ID==pic_id]
    lines=[]
    for i in range(len(pic)):
        t=pic.iloc[i]
        kind=mydic[t.NODE_TYPE]
        text = t.NODE_COORDINATE
        numbers = re.findall(r'\d+\.?\d*', text)
        numbers = [eval(n) for n in numbers]  
        w_ = numbers[2]
        h_ = numbers[3]
        numbers[2]=numbers[2]+numbers[0]
        numbers[3]=numbers[3]+numbers[1]
        x_ = (numbers[2]+numbers[0]) / 2
        y_ = (numbers[3]+numbers[1]) / 2
        lines.append("{} {} {} {} {}\n".format(kind,x_,y_,w_,h_))
#         print(kind,x_,y_,w_,h_)
    with open('1-200/labels/'+ '0000' + format(str(pic_id), '0>3s') + '.txt', 'w') as f:    
        f.writelines(lines)
    print(pic_id)

```
